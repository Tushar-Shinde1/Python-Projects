```python
dataset=pd.read_csv("/kaggle/input/execution-data-in-us/DPIC Execution Database - U.S. Executions.csv")
dataset
```

```python
dataset.fillna(10)
```

```python
# LabelEncoder
```

```python
from sklearn.preprocessing import LabelEncoder
l1=LabelEncoder()
```

```python
dataset["Execution Volunteer"]=l1.fit_transform(dataset["Execution Volunteer"])
dataset["Juvenile"]=l1.fit_transform(dataset["Juvenile"])
dataset["First Name"]=l1.fit_transform(dataset["First Name"])
dataset["Last Name"]=l1.fit_transform(dataset["Last Name"])
dataset["Middle Name(s)"]=l1.fit_transform(dataset["Middle Name(s)"])
dataset["Suffix"]=l1.fit_transform(dataset["Suffix"])
dataset["Race"]=l1.fit_transform(dataset["Race"])
dataset["Sex"]=l1.fit_transform(dataset["Sex"])
dataset["Region"]=l1.fit_transform(dataset["Region"])
dataset["Country"]=l1.fit_transform(dataset["Country"])
dataset["State"]=l1.fit_transform(dataset["State"])
dataset["Foreign National"]=l1.fit_transform(dataset["Foreign National"])
dataset["Execution Date"]=l1.fit_transform(dataset["Execution Date"])
dataset["Victim(s) Race(s)"]=l1.fit_transform(dataset["Victim(s) Race(s)"])
dataset
```

```python
dataset.corr()
```

```python
import seaborn as sns
sns.heatmap(dataset.corr())
```

```python
dataset["Execution Volunteer"].value_counts()
```

```python
x=dataset.iloc[:,1:].values
y=dataset.iloc[:,:1].values
```

```python
x
```

```python
y
```

```python
# SMOTE
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE()
x_data,y_data=sm.fit_resample(x,y)
```

```python
from collections import Counter
Counter(y_data)
```

```python
# StandardScaler
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_scaled=ss.fit_transform(x_data)
print(x_scaled)
```

```python
a=x_scaled.mean()
a=round(a)
a
```

```python
b=x_scaled.var()
b=round(b)
b
```

```python
# DATA_train_test_split
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y_data,test_size=0.2,random_state=9)
```

```python
# KNN
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_pred)*100
ac
```

```python
# SVM
```

```python
from sklearn.svm import SVC
svm = SVC(kernel="linear")
svm.fit(x_train,y_train)
```

```python
y_pred2=svm.predict(x_test)
y_pred2
```

```python
from sklearn.metrics import accuracy_score
ac2=accuracy_score(y_test,y_pred2)*100
ac2
```

```python
# Naive_bayes
```

```python
from sklearn.naive_bayes import GaussianNB
na=GaussianNB()
na.fit(x_train,y_train)
```

```python
y_pred3=na.predict(x_test)
y_pred3
```

```python
from sklearn.metrics import accuracy_score
ac3=accuracy_score(y_test,y_pred3)*100
ac3
```

```python
# EnsembleTechnique
```

```python
from sklearn.ensemble import VotingClassifier
ab3=VotingClassifier(estimators=[("naive_bayes",na),("Support Vector Machine",svm),("K-NearestNeighbors",knn)])
ab3
```

```python
ab3.fit(x_train,y_train)
```

```python
y_pred_x=ab3.predict(x_test)
y_pred_x
```

```python
from sklearn.metrics import accuracy_score
ac4=accuracy_score(y_test,y_pred_x)*100
ac4
```

```python
import pandas as pd
import numpy as np
```

# Loading Training dataset

```python
dataset=pd.read_csv("/kaggle/input/titanic/train.csv")
dataset
```

# Data Pre-processing

```python
dataset.info()
```

```python
del dataset["Cabin"]
del dataset["Ticket"]
del dataset["Name"]
del dataset["PassengerId"]

dataset
```

```python
dataset["Age"].mean()
```

```python
dataset["Embarked"].value_counts()
```

```python
dataset["Age"]=dataset["Age"].fillna(29)
dataset["Embarked"]=dataset["Embarked"].fillna('Q')
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

# Convert String data into Numeric data

```python
from sklearn.preprocessing import LabelEncoder
l1=LabelEncoder()
dataset["Sex"]=l1.fit_transform(dataset["Sex"])
dataset["Embarked"]=l1.fit_transform(dataset["Embarked"])
dataset
```

# Slicing Training data

```python
x=dataset.iloc[:,1:].values
y=dataset.iloc[:,[0]].values
print(x)
print(y)
```

```python
dataset["Survived"].value_counts()
```

# Data Imbalancing

```python
from imblearn.over_sampling import SMOTE
S1=SMOTE()
X_train,Y_train=S1.fit_resample(x,y)
```

```python
from collections import Counter
Counter(Y_train)
```

# Normalization

```python
from sklearn.preprocessing import MinMaxScaler
MMS=MinMaxScaler()
X=MMS.fit_transform(X_train)
print(X.min())
print(X.max())
print(X)
```
